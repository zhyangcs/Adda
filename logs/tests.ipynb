# 测试 send_prompt_n 函数的多响应能力

本测试旨在验证当前版本的 send_prompt_n 函数是否能从API获取多个响应。

```python
import sys
import os
sys.path.append('/home/lizhenyu/projects/autofe')

from src.llm.utils.llm_util import send_prompt_n
import termcolor
```

```python
# 测试1：尝试获取3个响应
print(termcolor.colored("测试1：请求3个响应", "cyan"))
try:
    responses = send_prompt_n(
        role_prompt="你是一个助手",
        user_prompt="请列举3个不同的机器学习算法，每个算法用一句话描述",
        n=3,
        temperature=0.9
    )
    print(f"响应数量: {len(responses)}")
    for i, response in enumerate(responses):
        print(f"\n--- 响应 {i+1} ---")
        print(response)
except Exception as e:
    print(f"错误: {e}")
```

```python
# 测试2：尝试获取4个响应（模拟CodeAgent的使用场景）
print(termcolor.colored("\n测试2：请求4个响应（CodeAgent场景）", "cyan"))
try:
    responses = send_prompt_n(
        role_prompt="你是一个Python编程助手",
        user_prompt="请写一个Python函数，计算两个数的平均值。给出4个不同的实现版本。",
        n=4,
        temperature=0.8
    )
    print(f"响应数量: {len(responses)}")
    for i, response in enumerate(responses):
        print(f"\n--- 代码版本 {i+1} ---")
        print(response)
except Exception as e:
    print(f"错误: {e}")
```

```python
# 测试3：API限制测试
print(termcolor.colored("\n测试3：API限制测试", "cyan"))
for n in [1, 2, 3, 4, 5]:
    try:
        print(f"\n请求 {n} 个响应:")
        responses = send_prompt_n(
            role_prompt="",
            user_prompt=f"请说'你好'",
            n=n,
            temperature=0.5
        )
        print(f"成功获取 {len(responses)} 个响应")
        
        # 检查响应是否相同
        unique_responses = set(responses)
        print(f"唯一响应数量: {len(unique_responses)}")
        
    except Exception as e:
        print(f"请求 {n} 个响应时出错: {e}")
```

```python
# 分析测试结果
print(termcolor.colored("\n=== 测试结果分析 ===", "cyan"))
print("如果所有测试都只返回1个响应，说明API不支持n参数")
print("如果能返回多个响应，说明API支持n参数，越界检查只是预防性措施")
print("\n建议：")
print("1. 如果API不支持n参数，需要修改send_prompt_n函数通过多次调用来模拟")
print("2. 如果API支持n参数但偶尔返回不足，当前越界检查是合适的")
print("3. 检查OpenAI API版本和配置是否正确")
```
